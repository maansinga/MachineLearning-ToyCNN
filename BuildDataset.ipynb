```python
import sys
import os
import random
from shutil import copyfile
```

```python
squares = os.listdir('./Squares')
circles = os.listdir('./Circles')
csv = open("data.csv","w")
```

```python

i = 0
while i<300:
    if random.randint(0,1000)%2 == 0:
        # is a circle
#         os.system("cp ./Circles/%s ./Data/%s.png"%(circles[random.randint(0,399)], "%04d"%(i)))
        copyfile("./Circles/%s"%(circles[random.randint(0,399)]), "./Data/%04d.png"%(i,))
        csv.write("1, 0\n")
    else:
        # is a square
#         os.system("cp ./Squares/%s ./Data/%s.png"%(squares[random.randint(0,399)], "%04d"%(i)))
        copyfile("./Squares/%s"%(circles[random.randint(0,399)]), "./Data/%04d.png"%(i,))
        csv.write("0, 1\n")
        
    i += 1
    print("%dth image selected"%(i,))
    
csv.close()
```
