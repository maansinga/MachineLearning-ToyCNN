```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
img = np.zeros((64,64), dtype=np.uint8)
```

```python
def draw_circle(img):
    shape=img.shape
    d = shape[0] if shape[0] < shape[1] else shape[1]
    r = int(d/2) - 3
    if r < 3:
        return False
    else:
        x = int(shape[0]/2)
        y = int(shape[1]/2)
        print(img.shape, r)
        cv2.circle(img, (x,y), r, color=255, thickness=2, lineType=cv2.LINE_AA)
        
        return True
```

```python
roi = img[10:img.shape[0]-10, 10:img.shape[1]-10]
draw_circle(roi)

fig = plt.figure(figsize=(10,10),)
plt.imshow(img, cmap='binary')
```

```python
def roi_generator(shape):
    if(shape[0]<=10 or shape[1]<=10):
        return False
    
    x0 = 0
    y0 = 0
    x1 = 0
    y1 = 0
    
    while x1-x0 < 6 or y1-y0 < 6:
        x0 = random.randint(0,shape[0]-1)
        y0 = random.randint(0,shape[1]-1)
        r = shape[0]-x0 if shape[0] - x0 < shape[1] - y0 else shape[1]-y0 
        x1 = x0 + r
        y1 = y0 + r
    return(((x0,y0),(x1,y1)))

roi_generator((20,20))
```

```python
i = 0
while i < 400:
    img = np.zeros((64,64))
    roi = roi_generator(img.shape)
    if draw_circle(img[roi[0][0]:roi[1][0], roi[0][1]:roi[1][1] ]) is True:
        print(roi)
        cv2.imwrite("./Circles/%0d.png"%(i,),img)
        print("Created %dth circle"%(i,))
        i += 1
    
    
```
